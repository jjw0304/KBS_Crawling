```python
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
chrome_driver_path = r'C:\Users\jjw03\Downloads\chromedriver.exe'  # chrome driver의 경로
driver = webdriver.Chrome(chrome_driver_path)
data = []
for i in range(1, 22):
    driver.get('http://news.kbs.co.kr/vod/program.do?bcd=0001#2019.04.' + str(i))
    html = driver.page_source
    soup = BeautifulSoup(html)
    table = soup.find('div', {'class' : 'c-inner newslist-grid'})
    for news_list in table.find_all('li'):
        news = news_list.find('em').text
        data.append(news)
```
